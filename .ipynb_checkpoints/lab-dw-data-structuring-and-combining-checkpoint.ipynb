```python
# Function to clean column names
def clean_column_names(columns):
    new_cols = []
    for col in columns:
        col = col.upper()                 
        col = col.strip()                 
        col = col.replace(" ", "_")         
        new_cols.append(col)
    return new_cols
```

```python
file_combined.columns = clean_column_names(file_combined.columns)
```

```python
file_combined
```

```python
# data_cleaning.py

def standardize_gender(file_combined, column="GENDER"):
    """Standardize Gender column to 'M' and 'F'."""
    if column in file_combined.columns:
        file_combined[column] = file_combined[column].str.upper().replace({
            "FEMALE": "F",
            "FEMAL": "F",
            "MALE": "M"
        })
    return file_combined
    
def standardize_states(file_combined, column="STATES"):
    """Replace state abbreviations with full names."""
    state_mapping = {
        "AZ": "Arizona",
        "Cali": "California",
        "WA": "Washington"
        }
    if column in file_combined.columns:
        file_combined[column] = file_combined[column].replace(state_mapping)
    return file_combined

def standardize_education(file_combined, column="EDUCATION"):
    """Standardize education column (Bachelors -> Bachelor)."""
    if column in file_combined.columns:
        file_combined[column] = file_combined[column].replace({
            "Bachelors": "Bachelor"
        })
    return file_combined

def clean_customer_lifetime_value(file_combined, column="CUSTOMER_LIFETIME_VALUE"):
    """Remove % or $ and convert to numeric."""
    if column in file_combined.columns:
        file_combined[column] = file_combined[column].astype(str).str.replace("%", "").str.replace("$", "").str.strip()
        file_combined[column] = pd.to_numeric(file_combined[column], errors='coerce')
    return file_combined

def standardize_vehicle_class(file_combined, column="VEHICLE_CLASS"):
    """Standardize vehicle class to simpler categories."""
    if column in file_combined.columns:
        file_combined[column] = file_combined[column].replace({
            "Sports Car": "Luxury",
            "Luxury SUV": "Luxury",
            "Luxury Car": "Luxury"
        })
    return file_combined

def clean_number_of_open_complaints(file_combined, column="NUMBER_OF_OPEN_COMPLAINTS"):
    """Extract middle value from strings like '1/5/00' and convert to numeric."""
    if column in file_combined.columns:
        def extract_middle(value):
            if pd.isna(value):
                return 0
            parts = str(value).split("/")
            return int(parts[1]) if len(parts) >= 2 else int(parts[0])
        file_combined[column] = file_combined[column].apply(extract_middle)
    return file_combined

def fill_missing_values(file_combined):
    """Fill missing numeric values with mean, categorical with mode."""
    num_cols = file_combined.select_dtypes(include=['int64', 'float64']).columns
    cat_cols = file_combined.select_dtypes(include=['object', 'category']).columns

    for col in num_cols:
        file_combined[col].fillna(file_combined[col].mean(), inplace=True)
    for col in cat_cols:
        file_combined[col].fillna(file_combined[col].mode()[0], inplace=True)
    return file_combined

def drop_duplicates(file_combined, subset=None):
    """Drop duplicate rows based on optional subset of columns."""
    file_combined.drop_duplicates(subset=subset, keep='first', inplace=True)
    file_combined.reset_index(drop=True, inplace=True)
    return file_combined
```

```python
def main(file_combined):
    """Main function to perform all cleaning and formatting steps."""
    file_combined = standardize_gender(file_combined)
    file_combined = standardize_states(file_combined)
    file_combined = standardize_education(file_combined)
    file_combined = clean_customer_lifetime_value(file_combined)
    file_combined = standardize_vehicle_class(file_combined)
    file_combined = clean_number_of_open_complaints(file_combined)
    file_combined = fill_missing_values(file_combined)
    file_combined = drop_duplicates(file_combined)
    return file_combined
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Create pivot table: total revenue by sales channel
pivot = pd.pivot_table(
    marketing_customer_analysis,
    index="SALES_CHANNEL",
    values="TOTAL_CLAIM_AMOUNT",
    aggfunc="sum"
).round(2)

print(pivot)
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Create the pivot table
pivot_clv = pd.pivot_table(
    marketing_customer_analysis,
    index='GENDER',           
    columns='EDUCATION',       
    values='CUSTOMER_LIFETIME_VALUE',  
    aggfunc='mean'             
).round(2)                     

print(pivot_clv)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
# Group by policy type and month, count complaints
summary = marketing_customer_analysis.groupby(['POLICY_TYPE', 'MONTH'])['NUMBER_OF_OPEN_COMPLAINTS'].sum().reset_index()
```

```python
# Rename column for clarity
summary = summary.rename(columns={'NUMBER_OF_OPEN_COMPLAINTS': 'TOTAL_COMPLAINTS'})
```
